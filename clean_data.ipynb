```python
import pandas as pd
path = 'superstore dataset/Sample - Superstore.csv'
data = pd.read_csv(path)
```

```python
df = data.copy()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
for date in df['order_date']:
    df['order_date'] = pd.to_datetime(date).strftime('%Y-%m-%d')    
print("All order dates have been converted to YYYY-MM-DD format.")

for date in df['ship_date']:
    df['ship_date'] = pd.to_datetime(date).strftime('%Y-%m-%d')
print("All ship dates have been converted to YYYY-MM-DD format.")
```

```python
import re

# Define a function to check for problematic characters
def check_problematic_chars(value):
    if isinstance(value, str):
        return bool(re.search(r"[\'\"\\]", value))
    return False

# Apply the function to the dataframe and get the problematic values
problematic_values = df.applymap(check_problematic_chars)

# Get the rows and columns with problematic values
rows_with_issues = problematic_values.any(axis=1)
columns_with_issues = problematic_values.any(axis=0)

# Display the problematic rows and columns
problematic_df = df.loc[rows_with_issues, columns_with_issues]
problematic_df
```
