```python
import pandas as pd
path = 'superstore dataset/Sample - Superstore.csv'
data = pd.read_csv(path)
```

```python
df = data.copy()
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
print("All order dates have been converted to a standard format of YYYY-MM-DD.")
df.head()
```

```python
df['ship_date'] = pd.to_datetime(df['ship_date'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
print("All ship dates have been converted to a standard format of YYYY-MM-DD.")
df.head()
```

```python
# Checking max length of each column to set SQL data types.
max_lengths = df.astype(str).map(len).max()
print(max_lengths)
```
